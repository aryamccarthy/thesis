```python
import networkx as nx
```

```python
G = nx.MultiGraph()
```

```python
G.add_edges_from([(1, 2), (1, 2), (1, 3), (1, 4), (1, 4), (2, 3), (3, 4)])
```

```python
nx.draw_networkx(G)
```
