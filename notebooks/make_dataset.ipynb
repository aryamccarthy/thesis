```python
import networkx as nx
from tqdm import tqdm
```

```python
nx.__version__
```

We'll use the parameters from [Yang et al.](http://www.nature.com/articles/srep30750/) for generating networks.

```python
AVERAGE_DEGREE = 20
DEGREE_DISTRIBUTION_EXPONENT=2.01
COMMUNITY_SIZE_DISTRIBUTION_EXPONENT=1.01

def maximum_degree(n):
    return int(0.1 * n)

def maximum_community_size(n):
    return int(0.1 * n)

N_MIXING_COEFFICIENTS = 10
MIN_MIXING_COEFFICIENT = 0.03
MAX_MIXING_COEFFICIENT = 0.75
MIXING_COEFFICIENTS = np.linspace(MIN_MIXING_COEFFICIENT,
                                  MAX_MIXING_COEFFICIENT,
                                  N_MIXING_COEFFICIENTS)

N_NETWORK_SIZES = 10
MIN_NETWORK_SIZE = 233
MAX_NETWORK_SIZE = 31948
NETWORK_SIZES = range(MIN_NETWORK_SIZE,
                     MAX_NETWORK_SIZE,
                     N_NETWORK_SIZES)
```

```python
for n in tqdm(NETWORK_SIZES):
    for mu in tqdm(MIXING_COEFFICIENTS):
        G = nx.LFR_benchmark_graph(n=n, tau1=DEGREE_DISTRIBUTION_EXPONENT, 
                                   tau2=COMMUNITY_SIZE_DISTRIBUTION_EXPONENT, mu=mu, 
                                   average_degree=AVERAGE_DEGREE,
                                   max_community=maximum_community_size(n),
                                   max_degree=maximum_degree(n), max_iters=1000,
                                   seed=1337)
```

```python
n = 250
tau1 = 3
tau2 = 1.5
mu = 0.1
G = nx.LFR_benchmark_graph(n, tau1, tau2, mu, average_degree=5,
                           min_community=20, seed=10)
```
